```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def aprox(theta,g):
    return -g*theta
def real(theta,g):
    return -np.sin(theta)*g
```

```python
theta=np.pi*5/180
g=9.8
l=2
v=0
t=0
ts=[]
xs=[]
dt=0.00001
for i in range(1000000):
    ts.append(t)
    xs.append(l*np.sin(theta))
    v=v+dt*real(theta,g)
    theta+=v*dt/l
    t+=dt

plot=plt.plot(ts,xs)
plt.show()
```

```python
theta=np.pi*5/180
g=9.8
l=2
v=0
t=0
ts=[]
xs=[]
dt=0.00001
for i in range(1000000):
    ts.append(t)
    xs.append(l*np.sin(theta))
    v=v+dt*aprox(theta,g)
    theta+=v*dt/l
    t+=dt
plot=plt.plot(ts,xs)
plt.show()
```

```python
theta=np.pi*5/180
theta_1=np.pi*5/180
g=9.8
l=2
v=0
v_1=0
t=0
t_1=0
ts=[]
xs=[]
ts_1=[]
xs_1=[]
dt=0.00001
for i in range(1000000):
    ts.append(t)
    xs.append(l*np.sin(theta))
    v=v+dt*real(theta,g)
    theta+=v*dt/l
    t+=dt
    
    
    ts_1.append(t_1)
    xs_1.append(l*np.sin(theta_1))
    v_1+=dt*aprox(theta_1,g)
    theta_1+=v_1*dt/l
    t_1+=dt
plt.plot(ts,xs,label='real')
plt.legend()
plt.show()
plt.plot(ts_1,xs_1,label='aprox')
plt.legend()
plt.show()

print (ts[-1]-ts_1[-1])

```
